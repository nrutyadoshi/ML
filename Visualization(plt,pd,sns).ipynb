Scatter Plot

```
fig, ax = plt.subplots() # figure and axis
ax.scatter(iris["sepal_length"],iris["sepal_width"])
ax.set_title("Iris")
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
```

```
colors = {"Iris-setosa":"r", "Iris-versicolor":"g", "Iris-virginica":"b"} # dictionary of colours
fig, ax = plt.subplots()
for i in range(len(iris["sepal_length"])):
  ax.scatter(iris["sepal_length"][i], iris["sepal_width"][i], color=colors[iris["species"][i]])
ax.set_title("Iris")
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
```

```
iris.plot.scatter(x='sepal_length',y='sepal_width',title='Iris')
```

```
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris)
```

```
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
```

Line Chart

```
columns = iris.columns.drop(['species']) # Y axis
x = range(0,iris.shape[0]) # X axis
fig, ax = plt.subplots()
for column in columns:
  ax.plot(x,iris[column],label=column)
ax.set_title("Iris")
ax.legend() # label box in the graph
```

```
iris.drop('species',1).plot.line(title='Iris')
```

```
sns.lineplot(data=iris.drop(['species'],axis=1))
```

```
sns.kdeplot(data=iris.drop(['species'],axis=1)) # if outliers in dataset
```

Histogram

```
fig,ax = plt.subplots()
ax.hist(wine['points']) # how many times something occurs
ax.set_title("Wine")
ax.set_xlabel("points")
ax.set_ylabel("Frequency")
```

```
wine['points'].plot.hist()
```

```
iris.plot.hist(subplots=True, layout=(2,2), figsize=(10,10), bins=20)
```

```
sns.distplot(wine['points'],bins=10,kde=True)
```

Bar chart

```
fig,ax = plt.subplots()
data = wine['points'].value_counts() # frequency
points = data.index
frequency = data.values
ax.bar(points,frequency)
ax.set_title("Wine")
ax.set_xlabel("points")
ax.set_ylabel("Frequency")
```

```
wine['points'].value_counts().sort_index().plot.bar()
```

```
wine['points'].value_counts().sort_index().plot.barh() # horizontal
```

```
wine.groupby("country").price.mean().sort_values(ascending=False)[:5].plot.bar() # first 5 highest displayed based on values
```

```
iris.groupby("species").mean().sort_values(by="species").plot.bar(subplots=True, layout=(2,2), figsize=(10,10))
```

```
sns.countplot(wine['points'])
```

Box plot

```
wine_new = wine[(wine['points']>=95) & (wine['price']<=1000)]
sns.boxplot('points','price',data=wine_new)
```

Heatmap

```
plt.figure(figsize=(10,10))
sns.heatmap(iris.corr(),annot=True)
```

Faceting

```
f = sns.FacetGrid(iris, col="species")
f = f.map(sns.kdeplot, 'sepal_length')
```

Pairplot

```
sns.pairplot(iris)
```
